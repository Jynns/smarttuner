```python
import pyaudio as pa
import wave 
import struct
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 
import time as t

```

```python
FORMAT =  pa.paInt16
CHANNELS = 2
RATE = 44100
CHUNK = 1024*100
```

```python
p = pa.PyAudio()
```

```python
stream = p.open(format = FORMAT,
                channels = CHANNELS,
                rate = RATE,
                input = True)
```

```python
frames = []

frequency = 385.6
```

```python
for i in range(0, int(RATE / CHUNK)):
    data = stream.read(CHUNK)
    data_converted = struct.unpack(str(4 * CHUNK) + 'B',data)
    data_np = np.array(data_converted, dtype='b')[::2] + 128
    frames.append(data_np)
```

```python
fig, ax = plt.subplots(1, figsize=(15,7))
ax.set_xlim(2*CHUNK)
for i in range(20):
    ax.plot(frames[i])
    plt.show()
    t.sleep(5)
```
